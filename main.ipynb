```python
# Import modeling libraries
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

```

```python
# --- 2. Feature Engineering & Selection ---
print("\n--- Starting Feature Engineering ---")

def feature_engineer(df):
    """Applies feature engineering to the dataset."""
    # Convert startdate to datetime objects
    df['startdate'] = pd.to_datetime(df['startdate'],'dd,mm,yy')
    
    # Date-based features
    df['day_of_year'] = df['startdate'].dt.dayofyear
    df['month'] = df['startdate'].dt.month
    
    # Create cyclical features for day of year to capture seasonality
    # This is more effective for tree-based models than a simple linear day number
    df['sin_day_of_year'] = np.sin(2 * np.pi * df['day_of_year'] / 365.25)
    df['cos_day_of_year'] = np.cos(2 * np.pi * df['day_of_year'] / 365.25)
    
    # One-hot encode the climate region category
    # Using pd.get_dummies handles unseen categories in the test set gracefully
    df = pd.get_dummies(df, columns=['climateregions__climateregion'], prefix='climate', dtype=int)
    
    return df

train_df = feature_engineer(train_df)
test_df = feature_engineer(test_df)

print("Feature engineering complete.")
```

```python
# --- 3. Define Feature Set ---
# We will be selective and choose features that are most likely to be predictive.

# Core geographic and elevation features
geo_features = ['lat', 'lon', 'elevation__elevation']

# All 'contest-' prefixed features, which are direct inputs for the challenge
contest_features = [col for col in train_df.columns if 'contest-' in col and col != 'contest-tmp2m-14d__tmp2m']

# Mean of the NMME model forecasts (these aggregate multiple models, reducing noise)
nmme_mean_features = [col for col in train_df.columns if 'nmmemean' in col]

# Climate indices
climate_indices = ['mjo1d__phase', 'mjo1d__amplitude', 'mei__mei', 'mei__meirank', 'mei__nip']

# Engineered date and climate region features
engineered_features = [col for col in train_df.columns if 'sin_day' in col or 'cos_day' in col or 'climate_' in col]
engineered_features.append('month')


# Combine all selected features into our final feature list
features = geo_features + contest_features + nmme_mean_features + climate_indices + engineered_features

# The target variable we want to predict
target = 'contest-tmp2m-14d__tmp2m'

# Align columns between training and testing sets - crucial for prediction
# This adds columns with 0s to the test set if a climate region was only in the training set
train_labels = train_df[target]
train_ids = train_df['index']
test_ids = test_df['index']

# Drop columns not in our feature list or that we want to exclude
cols_to_drop = ['index', 'startdate', 'day_of_year', target]
train_df = train_df.drop(columns=cols_to_drop, errors='ignore')
test_df = test_df.drop(columns=cols_to_drop, errors='ignore')

# Align columns - crucial for consistent model input
# This ensures both dataframes have the same columns in the same order
common_cols = list(set(train_df.columns) & set(test_df.columns))
features = [f for f in features if f in common_cols]

X = train_df[features]
X_test = test_df[features]
y = train_labels

print(f"\nTraining with {len(features)} features.")


# --- 4. Model Training ---
print("\n--- Training LightGBM Model ---")

# Split data for validation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42)

# LightGBM is efficient and powerful for tabular data
lgb_params = {
    'objective': 'regression_l1',
    'metric': 'rmse',
    'n_estimators': 2000,
    'learning_rate': 0.01,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 1,
    'lambda_l1': 0.1,
    'lambda_l2': 0.1,
    'num_leaves': 32,
    'verbose': -1,
    'n_jobs': -1,
    'seed': 42,
    'boosting_type': 'gbdt',
}

model = lgb.LGBMRegressor(**lgb_params)

# Train the model with early stopping to prevent overfitting
model.fit(X_train, y_train,
          eval_set=[(X_val, y_val)],
          eval_metric='rmse',
          callbacks=[lgb.early_stopping(100, verbose=True)])


# --- 5. Prediction and Submission ---
print("\n--- Generating predictions on test data ---")
predictions = model.predict(X_test)

# Create submission file in the specified format
submission_df = pd.DataFrame({'index': test_ids, 'contest-tmp2m-14d__tmp2m': predictions})
submission_df.to_csv('submission.csv', index=False)

print("\n'submission.csv' created successfully!")
print("Submission file head:")
print(submission_df.head())
```
